```python
#import all the necessary dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load csv data
df=pd.read_csv(r"D:\subjects\year_2_sem_4_Data_Science\Python-int 375\Fake news detection\RockVsMineProject\rock_mine_dataset.csv",header=None)
```

```python
#print head
df.head()
```

```python
#check the shape of the given data
df.shape
```

```python
df[60].value_counts()
```

```python
# remem ber 0 represnt Rock and 1 represents mines
#seperating data for recognition
x_df=df.drop(columns=60,axis=1)
y_df=df[60]
print(x_df)
print(y_df)
```

```python
#to train the model
x_train,x_test,y_train,y_test=train_test_split(x_df,y_df,test_size=0.2,stratify=y_df,random_state=1)
```

```python
print(x_df.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
print(y_train)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#now to find accuracy
x_train_prediction=model.predict(x_train)
train_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on training data: ",train_data_accuracy)
```

```python
#test prediction
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy)
```
